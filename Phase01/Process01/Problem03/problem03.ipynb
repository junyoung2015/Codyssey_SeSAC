# 문제3: 추천시스템의 응용 분야와 사례

## 수행목표

- 추천시스템 응용 분야를 조사하고 사례를 정리한다.

## 수행단계

- 추천시스템의 주요 응용 분야를 5개 이상 조사해서 설명하라.
- 국내 주요 사이트의 추천시스템 활용 현황을 3개 이상 조사해서 설명하라.

## 추천시스템의 주요 응용 분야

### 1) 전자상거래 (eCommerce)

- **핵심 아이디어**  
  사용자의 **구매 이력** 및 **브라우징 히스토리** 등을 바탕으로, 맞춤형 상품을 추천하여 매출 상승에 기여
- **사례**: 아마존(Amazon)  
  - “이 상품을 구매한 고객이 함께 본 상품” 또는 “이 상품을 본 고객이 결국 구매한 상품” 등 추천
  - 맥킨지(McKinsey)에 따르면, 아마존 전체 매출의 약 35%가 추천 시스템 덕분에 발생했다고 함
- **효과**  
  - 고객이 원하는 상품을 빠르게 찾을 수 있어 만족도 및 체류 시간 증가
  - 교차 판매(Cross-selling)와 업셀(Up-selling) 가능

### 2) 엔터테인먼트/미디어 스트리밍 (OTT)

- **핵심 아이디어**  
    - 시청(또는 청취) 기록, 평점, 장르 선호 등을 분석해 **사용자에게 맞춤형 콘텐츠(영상, 음악)를 추천**
- **사례**: 넷플릭스(Netflix), 스포티파이(Spotify)  
  - 넷플릭스: 협업 필터링(CF)과 콘텐츠 기반 필터링을 혼합하여, 개별 사용자에게 시청할 만한 (해당 사용자가 시청할 확률이 높은) 작품 추천
    - 사용자의 80%가 추천 받은 콘텐츠를 시청
  - 스포티파이: “Discover Weekly” 개인화 플레이리스트 생성
    - 청취 이력, 아티스트·곡 정보(NLP 분석) 등을 종합 활용
- **효과**  
  - 개인 취향에 따른 콘텐츠 제공으로 **이탈률 감소** 및 **서비스 만족도 증가**
  - 더 많은 콘텐츠를 탐색하게 하여 **플랫폼 체류 시간**과 **유료 구독 유지율**을 높이는데 쓰임

### 3) 소셜 미디어

- **핵심 아이디어**  
    - 사용자들의 **친구 관계**, **좋아요/댓글/공유** 활동 이력, **관심사 태그** 등을 종합적으로 분석해, **피드(Feed), 광고, 친구/그룹 추천** 등을 개인화
- **사례**: 메타(Meta: 페이스북·인스타그램), 링크드인(LinkedIn)
  - 페이스북/인스타그램: 사용자가 아직 팔로우하지 않는 콘텐츠나 크리에이터를 추천해 도달 범위를 확대
  - 링크드인: “당신이 이 사랑을 알 수도 있습니다” 와 같이 비즈니스/네트워크 확장에 도움을 주는 기능 제공  
- **효과**
  - 사용자의 **콘텐츠 탐색 범위 확대**, **맞춤 광고 노출**로 광고 수익 증대
  - **사용자 참여도(Engagement)** 향상 및 플랫폼 활성화

### 4) 헬스케어

- **핵심 아이디어**
    - 환자의 **의료 기록**, **생활 습관 데이터** 등을 기반으로, 맞춤형 진단/치료 정보를 제시
- **사례**: Ada Health 등
  - 사용자가 입력한 증상을 바탕으로 가능한 질환 목록과 **의료 조치**(병원 방문, 상담 등)를 추천
- **효과**  
  - 빠른 진단 가이드와 예방 정보 제공 → **의료비 절감**, **치료 효율성 상승**
  - 개별 환자의 상태에 맞춘 **맞춤형 건강관리**로 환자 만족도 제고

### 5) 여행/숙박 및 오프라인 리테일

- **핵심 아이디어**  
  - 사용자 **여행 목적지**, **과거 예약 이력**, **취향(호캉스 vs. 배낭여행)** 등을 분석하여 숙박·항공편·관광 상품 추천
- **사례**: Airbnb, Booking.com 등  
  - 예약 패턴 + 위치 기반 데이터 + 유사 취향 이용자 패턴 등을 종합해 “가장 적합한 숙소” 추천
- **효과**
  - 사용자가 **시간을 절약**하여 빠르게 예약할 수 있음 → 고객 편의성 향상
  - **객실 점유율**과 **부가서비스**(액티비티, 렌터카 등) 판매 상승

## 국내 주요 사이트의 추천시스템 활용 현황

### 1) Laftel – 애니메이션 OTT 서비스

1. **서비스 특징**
   - **애니메이션 스트리밍 플랫폼**으로, 사용자 취향에 따라 작품을 추천
   - 장르/태그/제작사/브랜드 등으로 분류된 방대한 DB를 통해 개인화된 애니 리스트를 제공
2. **추천 시스템 사용 방식**
   - **시청 이력**, **선호 장르** 정보를 기반으로 개별 사용자에게 적합한 작품 큐레이션
   - 단순 인기 순위가 아니라, 사용자의 시청패턴(완결작 선호, 특정 스튜디오 팬 등)을 파악해 맞춤 추천
      - 콘텐츠 기반 필터링: 작품의 장르, 태그, 제작사 등 메타데이터를 토대로 비슷한 작품을 추천
      - 협업 필터링: 유사한 취향을 가진 다른 사용자가 선호하는 애니메이션을 큐레이션
3. **성과 및 의의**
   - 불법 다운로드가 만연한 애니메이션 시장에서, **합법적 시청 문화** 정착에 기여
   - SVOD/TVOD/AVOD 등 다양한 모델을 통해 **개인 맞춤 구독 옵션**도 제안

### 2) Wanted – 네트워크 기반 AI 채용 매칭

1. **서비스 특징**
   - AI 기반 **직장/포지션 매칭** 및 HR 솔루션을 제공 (ZipRecruiter, Eightfold.ai, Upwork, Fiverr 등과 유사)
   - 구직자 정보를 기반으로 딥러닝 매칭 엔진을 적용하여 “맞춤형 채용 공고” 추천
2. **추천 시스템 사용 방식**
   - **구직자의 역량, 경력, 선호 산업/직무** 등을 파악 → 해당 조건과 유사한 기업/직무를 추천
   - 기업 입장에서도 **적합 후보**를 자동으로 추천받아 채용 프로세스 효율화
   - 유저/포지션 매칭결과 학습 후, 개별 유저/포지션 매칭결과 예측
3. **성과 및 의의**
   - 불필요한 이력서 열람/면접 부담을 줄여, **구인-구직 매칭 효율**을 높임
   - **인재-기업** 모두에게 맞춤형 선택지 제시로 **채용 성공률 개선**

### 3) Naver Shopping – 온라인 커머스 추천 & 검색 노출

1. **서비스 특징**
   - 국내 대표 포털 네이버의 온라인 쇼핑 영역에서, 검색어 입력 시 상품 노출 순위를 결정하는 **추천/랭킹 알고리즘** 활용
2. **추천 시스템 사용 방식**
   - **적합도(제품명/브랜드/카테고리/태그), 인기(클릭, 판매량, 리뷰 등), 신뢰도(SEO, 악의적 ‘어뷰징’ 방지)** 등의 복합 지표를 활용해 상단 노출 순위 결정
     - 콘텐츠 기반 + 지식 기반 필터링의 요소가 혼합된 형태
   - 사용자가 더 흥미를 느낄 만한 상품을 상위에 노출하여 **구매 전환**을 높임
3. **성과 및 의의**
   - 판매자(스마트스토어 운영자 등)들에게는 **상위 노출을 위한 가이드라인**(카테고리 정확성, 브랜드 정보 기입 등)을 제공
   - 사용자에게는 **보다 연관성 높은 상품 추천**을 통해, 쇼핑 탐색 시간을 줄이고 만족도 높임
